```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Result.csv')
df_test_1 = pd.read_csv('Instances/VCP/VCP-Test I.csv') 
df_test_2 = pd.read_csv('Instances/VCP/VCP-Test II.csv') 

df.head()
```

```python
frequency = df['INSTANCE'].value_counts()

print(frequency)
```

```python
df_res = df.copy()
data = {'Test1':{},'Test2':{}}
for i,value in enumerate(df_res["MODEL"].unique()):
    print(value)
    data['Test1']['HH-0' + str(i)] = list(df_res[(df_res['MODEL'] == value) & (df_res["TEST"] == 'Test I') & (df_res['SIZE'] == 'VCP-Training-00NAs')]['HH'])
    data['Test2']['HH-0' + str(i)] = list(df_res[(df_res['MODEL'] == value) & (df_res["TEST"] == 'Test II')& (df_res['SIZE'] == 'VCP-Training-00NAs')]['HH'])

print(data)
df_res_1 = pd.DataFrame.from_dict(data['Test1'])
df_res_2 = pd.DataFrame.from_dict(data['Test2'])

df_res_1.head()
```

```python
df_res_2.info()
```

```python
df_res_1.to_excel('Insights_test_1.xlsx','sheet_1')
df_res_2.to_excel('Insights_test_2.xlsx','sheet_1')
```

```python
df_test = pd.concat([df_test_1,df_test_2])
```

```python
df = pd.merge(df,df_test,on='INSTANCE', how='left')
df = df.drop(['DENSITY','MAX_DEG','MIN_DEG','BEST'],axis=1)
```

```python
features = ['DEF','DEG','COL_DEG','UNCOL_DEG','ORACLE']
for feature in features:

    df['MATCH_'+feature] = np.where(np.subtract(df['HH'],df[feature]) == 0,1,0)
    df['IMPROVEMENT_'+feature] = np.where(df['HH'] < df[feature],1,0)

df.head()
```

```python
res = df.drop(['INSTANCE'],axis=1).groupby(['MODEL','SIZE']).sum()
res
```

```python
res.to_excel('Insights.xlsx','sheet_1')
```
