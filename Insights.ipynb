```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Result.csv')
df_test_1 = pd.read_csv('Instances/VCP/VCP-Test I.csv') 
df_test_2 = pd.read_csv('Instances/VCP/VCP-Test II.csv') 

df.head()
```

```python
df_test = pd.concat([df_test_1,df_test_2])
```

```python
df = pd.merge(df,df_test,on='INSTANCE', how='left')
df = df.drop(['DENSITY','MAX_DEG','MIN_DEG','DEF','DEG','COL_DEG','UNCOL_DEG','BEST'],axis=1)
```

```python
df['MATCH'] = np.where(np.subtract(df['HH'],df['ORACLE']) == 0,1,0)
df['IMPROVEMENT'] = np.where(df['HH'] < df['ORACLE'],1,0)
df.head()
```

```python
res = df.drop(['INSTANCE'],axis=1).groupby(['MODEL','SIZE']).sum()
res
```

```python
res.to_excel('Insights.xlsx','sheet_1')
```
